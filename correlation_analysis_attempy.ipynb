```python
import scipy
from pyPPG import PPG, Fiducials, Biomarkers
from pyPPG.datahandling import load_data, plot_fiducials, save_data
import pyPPG.preproc as PP
import pyPPG.fiducials as FP
#import preprocessing.fiducials as FP
import pyPPG.biomarkers as BM
import pyPPG.ppg_sqi as SQI




from utils.dataloaders import OneSignal

import numpy as np
import sys
import json
import pandas as pd

data_path = "dataset/data/S001_128.mat" # the path of the file containing the PPG signal to be analysed
start_sig = 0 # the first sample of the signal to be analysed
end_sig = -1 # the last sample of the signal to be analysed (here a value of '-1' indicates the last sample)
savingfolder = 'temp_dir'
savingformat = 'csv'


# signal = load_data(data_path=data_path, start_sig=start_sig, end_sig=end_sig, use_tk=False)
# signal.v = signal.v [0:20*signal.fs]
signal = OneSignal("S001_128")
signal.filter()

s = signal.s
fiducials = signal.fiducials

fp = Fiducials(fp=fiducials) #fix this so that not directly from PPG package?

bmex = BM.BmCollection(s=s, fp=fp)


bm_defs, bm_vals, bm_stats = bmex.get_biomarkers()


bm = Biomarkers(bm_defs=bm_defs, bm_vals=bm_vals, bm_stats=bm_stats)

fp_new = Fiducials(fp.get_fp() + s.start_sig) 
save_data(s=s, fp=fp_new, bm=bm, savingformat=savingformat, savingfolder=savingfolder)

```

```python
df = pd.read_csv('temp_dir\Biomarker_vals\S001_128_ppg_sig_btwn_0--1.csv')
df.head()
```

```python
labels = scipy.io.loadmat('dataset\label\S001_128.mat')
label_col = labels['labels']
```

```python
print(label_col)
```

```python

pulse_indices = df["Index of pulse"].to_list()


pulse_labels = [label_col[i] for i in pulse_indices]

binary_labels = []
for i in pulse_labels:
    if(i == "N"):
        binary_labels.append(1)
    else:
        binary_labels.append(0)



df["labels"] = binary_labels
print(binary_labels)
```

```python
df.head()
```

```python
target = df['labels']

correlation_matrix = df.corrwith(target)

correlation_df = pd.DataFrame({'Feature': correlation_matrix.index, 'Correlation': correlation_matrix.values})

correlation_df['Absolute Correlation'] = correlation_df['Correlation'].abs()
correlation_df = correlation_df.sort_values(by='Absolute Correlation', ascending=False)

print(correlation_df)
```
